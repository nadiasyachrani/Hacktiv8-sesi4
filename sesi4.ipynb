# QUIZ 1
Buat function kalkulator dengan operator :
* 'x' sebagai kali
* ':' sebagai bagi
* '+' sebagai tambah
* '-' sebagai kurang
* '^' sebagai pangkat

```python
def calculator(a,b,c):
    if (a.isnumeric()==True) and (c.isnumeric()==True):
        a=int(a)
        c=int(c)
        if b=='x':
            print('Hasil dari ', a,b,c, 'adalah ', a*c)
        elif b=='+':
            print('Hasil dari ', a,b,c, 'adalah ', a+c)
        elif b==':':
            print('Hasil dari ', a,b,c, 'adalah ', a/c)
        elif b=='-':
            print('Hasil dari ', a,b,c, 'adalah ', a-c)
        elif b=='^':
            print('Hasil dari ', a,b,c, 'adalah ', a**c)
        else:
            print('operator', b, 'tidak ditemukan')
    else:
        print('salah')

a=(input('masukkan angka1= '))
b=(input('masukkan operator= '))
c=(input('masukkan angka2= '))

calculator(a,b,c)
```

```python
def kalkulator():
    try:
        angka1=float(input('masukkan angka pertama : '))
        operasi=input('masukkan operasi (+, -, x, :, ^) : ')
        angka2=float(input('masukkan angka kedua : '))
        
        if operasi=='+':
            hasil=angka1+angka2
        elif operasi=='-':
            hasil=angka1-angka2
        elif operasi=='x':
            hasil=angka1*angka2
        elif operasi==':':
            hasil=angka1/angka2
        elif operasi=='^':
            hasil=angka1**angka2
        else:
            print("operasi tidak valid")
            return
        
        print(f"hasil dari {angka1}{operasi}{angka2}={hasil}")

    except ValueError:
        print("input tidak valid, pastikan anda memasukkan angka")
        
kalkulator()
```

```python
tinggi = int(input('How Tall? '))
if tinggi%2 == 0:
    tinggi -= 1
tengah = int(tinggi/2) + 1

z = ''
for i in range(tengah):
    z += '  '*i+'* '*(tinggi-i*2)+'\n'
for i in range(tengah-1):
    j = tengah-1 - (i+1)
    z += '  '*j+'* '*(tinggi-j*2)+'\n'
print(z)
```

```python
def kelompok_3(tinggi):
    if tinggi%2 == 0:
        tinggi -= 1
    tengah = int(tinggi/2) + 1

    z = ''
    for i in range(tengah):
        z += '  '*i+'* '*(tinggi-i*2)+'\n'
    for i in range(tengah-1):
        j = tengah-1 - (i+1)
        z += '  '*j+'* '*(tinggi-j*2)+'\n'
    return z
a = int(input('How Tall? '))
print(kelompok_3(a))
```

# QUIZ 3
buat prime number generator based on interval
* input : 
* - please, enter the lower range value :
* - please, enter the upper range value : 
* output : 
* - the prime numbers in the range are : 

```python
lower=int(input('please, enter the lower range value : '))
upper=int(input('please, enter the upper range value : '))
print("Prime numbers between", lower, "and", upper, "are:")

for number in range(lower, upper + 1):
   # all prime numbers are greater than 1
   if number > 1:
        for i in range(2, number):
            if (number % i) == 0:
                break
        else:
            print(number)
```

```python
def is_prime(n):
    
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


def print_primes(start, end):
  
    for i in range(start, end + 1):
        if is_prime(i):
            print(i)

start = int(input("Masukkan nilai terkecil: "))
end = int(input("Masukkan nilai terbesar: "))

print("Bilangan prima antara", start, "dan", end, "adalah:")
print_primes(start, end)
```
